### Stock: Información del inventario por punto de venta, producto y por cadena. 

```python
import pandas as pd
import numpy as np
```

```python
# Importo el Dataframe en formato Excel
df_stock = pd.read_excel("stock.xlsx")  
```

```python
# Puedo ver una tupla con el número de filas y columnas del DataFrame
df_stock
```

```python
# Observo el tipo de datos que tiene el DataFrame
df_stock.dtypes
```

```python
# Con el comando info() me hago una idea con los campos que voy a trabajar
df_stock.info()
```

```python
# Creo una variable para saber si existen valores nulos en el dataframe 
valores_nulos_stock = df_stock.isnull()
valores_nulos_stock
```

```python
# Observo que existen valores nulos en la columna Mínimo, lo cual se tomará en cuenta en caso de ser relevante para el análisis.
valores_nulos_stock.sum()
```

```python
# Con describe() puedo obtener un análisis descriptivo del Dataframe que puede ser útil. 
df_stock.describe()
```

```python
# con any() verifico si hay filas duplicadas en el DataFrame
hay_duplicados = df_stock.duplicated().any()

if hay_duplicados:
    print("Existen filas duplicadas en el DataFrame.")
else:
    print("No hay filas duplicadas en el DataFrame.")
```
